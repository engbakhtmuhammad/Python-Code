```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image = cv2.imread('images/cat_grey.jpg', cv2.IMREAD_GRAYSCALE)

# Quantization function
def quantize_image(image, levels):
    # Calculate the mapping factor
    mapping_factor = 256 / levels
    # Apply quantization
    quantized_image = (image / mapping_factor).astype(np.uint8)
    # Scale back to original intensity range

    quantized_image = quantized_image * mapping_factor
    return quantized_image

# Original image
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

# Quantize image to 32 levels
quantized_image_32 = quantize_image(image, 32)
plt.subplot(1, 3, 2)
plt.imshow(quantized_image_32, cmap='gray')
plt.title('Quantized to 32 levels')

# Quantize image to 16 levels
quantized_image_16 = quantize_image(image, 16)
plt.subplot(1, 3, 3)
plt.imshow(quantized_image_16, cmap='gray')
plt.title('Quantized to 16 levels')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the grayscale medical image
image = cv2.imread('images/xray_grey.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Otsu's thresholding
_, binary_image = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display the original and binary images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image (Segmented ROIs)')

plt.tight_layout()
plt.show()
```
