```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
cat_image = cv2.imread("images/parrots.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Histogram equalization
equalized_cat_image = cv2.equalizeHist(cat_image)

# Calculate histograms
hist_cat_original = cv2.calcHist([cat_image], [0], None, [256], [0, 256])
hist_cat_equalized = cv2.calcHist([equalized_cat_image], [0], None, [256], [0, 256])

# Plotting
plt.figure(figsize=(12, 6))

# Original Image and Histogram
plt.subplot(2, 2, 1)
plt.imshow(cat_image, cmap="gray")
plt.title("Original Cat Image")

plt.subplot(2, 2, 2)
plt.plot(hist_cat_original, color="black")
plt.title("Original Cat Image Histogram")

# Equalized Image and Histogram
plt.subplot(2, 2, 3)
plt.imshow(equalized_cat_image, cmap="gray")
```
